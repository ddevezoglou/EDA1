```python
#Find the total courses per subject
df['subject'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
sns.countplot(data=df,x='subject',hue='subject')
plt.xticks(rotation=45);
plt.xlabel('Subjects');
plt.ylabel('Number of Courses');
```

```python
df['level'].value_counts()
```

```python
#Find total courses per level 
sns.countplot(data=df,x='level',hue='level')
plt.xticks(rotation=45);
plt.xlabel('Levels');
plt.ylabel('Number of Courses');
```

```python
df['is_paid'].value_counts()
```

```python
#Find total courses per level 
sns.countplot(data=df,x='is_paid',hue='is_paid')
plt.xticks(rotation=45);
plt.xlabel('Is_Paid');
plt.ylabel('Number of Courses');
plt.show()
```

```python
#Which courses has more number of lectures (free or paid )
df.groupby('is_paid')['num_lectures'].mean()
#The paid ones 
```

```python
#Which courses have a higher number of subscribers (free or paid )
df.groupby('is_paid')['num_subscribers'].mean()
#The free ones
```

```python
sns.barplot(data=df,x='is_paid',y='num_subscribers',hue='is_paid')
```

```python
#Which level has the highest number of subscribers
df.groupby('level')['num_subscribers'].sum()
```

```python
sns.barplot(data=df,x='level',y='num_subscribers',hue='level')
plt.xticks(rotation=50);
#';' is the same as plt.show() in this situation 
```

```python
#find most popular course title
df[df['num_subscribers']==df['num_subscribers'].max()]['course_title']
```

```python
#10 most popular courses as per number of subscribers
df.sort_values('num_subscribers',ascending=False)['course_title'].head(10)
```

```python
sns.barplot(x='num_subscribers',y='course_title',data=df.sort_values('num_subscribers',ascending=False).head(10),hue='course_title')
plt.show()
```

```python
#Find the course which is having the highest number of reviews 
df[df['num_reviews']==df['num_reviews'].max()]['course_title']
```

```python
#Most popular subject by number of reviews
sns.barplot(data=df,x='subject',y='num_reviews',hue='subject')
plt.xticks(rotation=45)
plt.show()
```

```python
#Does price affect number of reviews 
sns.scatterplot(x='price',y='num_reviews',data=df);
#the more expensive the course the more reviews it has
```

```python
#Find total number of courses related to python 
df['course_title'].str.contains('Python',case=False).sum()
```

```python
#display top 10 most popular python courses asper number of subscribers
```

```python
df[df['course_title'].str.contains('Python',case=False)].sort_values('num_subscribers',ascending=False)['course_title'].head(10)
```

```python

sns.barplot(x='num_subscribers',y='course_title',data=df[df['course_title'].str.contains('Python',case=False)].sort_values('num_subscribers',ascending=False).head(10),hue='course_title')
```

```python
#in which year the highest number of courses we posted?
df['year']=df['published_timestamp'].dt.year
```

```python
sns.countplot(data=df,x='year',hue='year',palette='tab10');
```

```python
df['year'].value_counts().head(1)
```

```python
#Display categorywise count of posted subjects (year wise)
df.groupby('year')['subject'].value_counts()
```
