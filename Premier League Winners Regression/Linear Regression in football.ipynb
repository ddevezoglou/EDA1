```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("Standings.csv")
```

```python
#As goal difference
df['GD']=df['GF']-df['GA']
```

```python
y=df.Rk
x=df.GD.values.reshape(-1,1)
```

```python
print(x.shape,y.shape)
```

```python
model=LinearRegression().fit(x,y)
```

```python
r_sq=model.score(x,y)
intercept=model.intercept_
slope=model.coef_
```

```python
r_sq
```

```python
y_pred=intercept+slope*x
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,c='red')
plt.ylim(-5,20.5)
plt.yticks([x for x in range(1,21)])
plt.gca().invert_yaxis()
plt.xlabel('Goal differential')
plt.ylabel('League Rank')
plt.title('Evaluating the Relationship between goal differential and League rank')
plt.savefig("graph.png")
```
