```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#Correlation matrix 
df.corr()
```

```python
#How many people have heart disease,and how many dont have 
df.columns
```

```python
df['target'].value_counts()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x=df['target'],hue=df['target'],legend=False,width=0.4)
plt.xticks([0,1],["Don't have h.d.",'Have h.d.'])
```

```python
#find count of male adn female in dataset 
df.columns
```

```python
df['sex'].value_counts()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x=df['sex'],hue=df['sex'],legend=False,width=0.4)
plt.xticks([0,1],['Female','Male']);
```

```python
#find the gender distribution 
sns.countplot(data=df,x='sex',hue='target',palette='tab10')
plt.xticks([1,0],['Male','Female'])
plt.legend(labels=['No Disease','Disease'])
plt.show()
```

```python
df.pivot_table(index='target',columns='sex',aggfunc='count',values='age')
```

```python
#age distribution 
plt.figure(figsize=(7,5))
sns.kdeplot(df['age'],fill=True,multiple='layer')
```

```python
#Check chest pain type
sns.countplot(data=df,x='cp',hue='cp',legend=False,palette='tab10',width=0.6)
plt.xticks([0,1,2,3],['typical angina','atypical angina','non anginal pain','asymptomatic'],rotation=15)
plt.xlabel('Chest Pain')
plt.ylabel('Count')
plt.show()
```

```python
#Show the chest pain distribution as per target variable
df.pivot_table(index='cp',columns='target',aggfunc='count',values='age')
```

```python
sns.countplot(data=df,x='cp',hue='target')
plt.xticks([0,1,2,3],['typical angina','atypical angina','non anginal pain','asymptomatic'],rotation=15)
plt.legend(['Non Disease','Disease'])
plt.xlabel('Chest Pain')
plt.show()
```

```python
#Show Fasting Blood Sugar Distribution according to target variable
df.columns
```

```python
df.pivot_table(index='fbs',columns='target',aggfunc='count',values='age')
```

```python
sns.countplot(data=df,x='fbs',hue='target')
plt.xticks([0,1],['No Diabetes', 'Diabetes'])
plt.legend(['Non Disease','Disease'])
plt.xlabel('Fasting Blood Sugar')
plt.show()
```

```python
#blood pressure distribution 
df['trestbps'].hist()
```

```python
#Compare RBP as per sex column 
g=sns.FacetGrid(df,hue='sex',aspect=3)
g.map(sns.kdeplot,'trestbps',fill=True)
plt.legend(labels=['Male','Female'])
plt.xlabel('Test Bloob Pressure')
plt.show()
```

```python
#distribution of serum cholesterol
sns.kdeplot(data=df,x='chol',fill=True)
```

```python
#plot continuous variables
cate_val=[]
cont_val=[]


for column in df.columns:
    if df[column].nunique()<=10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```
